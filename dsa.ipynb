```python
def rn(n):
    sign= -1 if n <0 else 1
    n=abs(n)
    rev=0
    while n>0:
        rev= rev*10+n%10
        n//=10
    return sign * rev
print(rn(-987))
```

```python
def tn(nums):
    t=0
    for n in nums:
        t+=n
    return t

print(tn([3,4,5,3]))
```

```python
def ev(nums):
    return [num for num in nums if num%2==0]
print(ev([1,2,3,4,5,6]))
```

```python
def freq(nums):
    d={}
    for n in nums:
        d[n]=d.get(n,0)+1
    return d

print(freq([2,3,43,3,5]))
print(freq(sorted([2,3,43,3,5])))
```

```python
def rev(s):
    return " ".join(s.split()[::-1])

print(rev("hello world python"))

def vd(d):
    return d[::-1]

print(vd("hello world python"))
    
```

```python
def ce(a,b,c):
    return list(set(a)&set(b)&set(c))

print(ce([1,2,3,4], [2,3,5], [2,3,6]))
```

```python
def cd(n):
    c=0
    while n>0:
        c+=1
        n//=10
    return c

print(cd(3245))
```

```python
def co(nums,x):
    return str(nums).count(str(x))

print(co(2334,9))
```

```python
def rs(s,k):
    k%=len(s)
    return s[-k:]+s[:-k]

print(rs("ritessd",4))
```

```python
def cap(s):
    return " ".join(word.capitalize() for  word in s.split())

print(cap("hj g"))
```

```python
def f_d(nums):
    seen=set()
    for num in nums:
        if num in seen:
            return num
        seen.add(num)
    return None

print(f_d([2, 3, 4, 4, 3]))
```

```python
def ss(nums):
    nums=list(set(nums))
    nums.sort()
    return nums[1]

print(ss([2,4,5,6,1,3]))
```

```python
def cw(s):
    return len(s.split())

print(cw("dfddf s  s "))
```

```python
def po2(n):
    return n>0 and (n&(n-1))==0

print(po2(4))
```

```python
import math
def power_of_two(n):
    if n > 0 and (n & (n - 1)) == 0:
        return int(math.log2(n))
    return -1

print(power_of_two(16))
```

```python
def single_num(nums):
    res = 0
    for num in nums:
        res ^= num
    return res

print(single_num([2,3,2,4,4])   )  
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car starts with key")

class Bike(Vehicle):
    def start(self):
        print("Bike starts with kick")

c = Car()
c.start()
```

```python
def mz(arr):
    j=0
    for i in range(len(arr)):
        if arr[i]!=0:
            arr[j],arr[i]=arr[i],arr[j]
            j+=1
    return arr
    
print(mz([2,2,3,4,0,3,0,3]))
```
